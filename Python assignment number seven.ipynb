Answer for Q.2. :-

Difference between the following operators:

(i) / and //:

The / operator performs floating-point division, resulting in a float value as the result.
The // operator performs floor division, resulting in an integer value as the result by discarding the decimal part.
Example:

python
Copy code
a = 10
b = 3

result1 = a / b  # Floating-point division
result2 = a // b  # Floor division

print(result1)  # Output: 3.3333333333333335
print(result2)  # Output: 3
(ii) ** and ^:

The ** operator is used for exponentiation in Python, raising a number to the power of another number.
The ^ operator is not used for exponentiation in Python. Instead, it is the bitwise XOR operator.
Example:

python
Copy code
a = 2
b = 3

result1 = a ** b  # Exponentiation
result2 = a ^ b  # Bitwise XOR

print(result1)  # Output: 8
print(result2)  # Output: 1

Right shift (>>) and left shift (<<) operators:

Right shift (>>): The right shift operator shifts the binary representation of a number to the right by a specified number of positions. It discards the rightmost bits and fills the vacant leftmost bits with the sign bit (for signed integers) or with zeros (for unsigned integers).
Example:

python
Copy code
num = 10  # Binary representation: 00001010

result = num >> 2  # Right shift by 2 positions

print(result)  # Output: 2 (Binary representation: 00000010)
Left shift (<<): The left shift operator shifts the binary representation of a number to the left by a specified number of positions. It fills the vacant rightmost bits with zeros.
Example:

python
Copy code
num = 10  # Binary representation: 00001010

result = num << 2  # Left shift by 2 positions

print(result)  # Output: 40 (Binary representation: 00101000)
